<h3 style="color:Red;"> 2. Data Pre-Processing</h3>

<h7 style="color:Black;"> ‚û°Ô∏è Performing neccessary cleaning steps:</h7>

```python
# Create a working copy for cleaning
df_clean = df.copy()
```

<h4 style="color:Black;">‚û°Ô∏èRemoving duplicates</h4>

```python
# Check duplicates
print("Number of duplicate rows before removal:", df_clean.duplicated().sum())
 
# Remove duplicates
df_clean.drop_duplicates(inplace=True)
 
print("Shape after removing duplicates:", df_clean.shape)
```

<h4 style="color:Black;">‚û°Ô∏èCapitalize first letter of all column names</h4>

```python
#To capitalize the first letter of every column name and replacing underscores with spaces
df.columns = df.columns.str.replace('_', ' ').str.title()

# To get Column Names
df.columns
```

<h4 style="color:Black;">‚û°Ô∏èHandling Missing Values</h4>

```python
# Check missing values
print("Missing values per column BEFORE handling:")
print(df_clean.isnull().sum())
```

```python
# a) Fill missing text fields with 'Unknown'
text_cols = [
    'language', 'text_content', 'mentions','emotion_type',
]
 
for col in text_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Unknown')
```

```python
# b) Fill numeric fields with median (or 0 if you prefer)
num_cols = [
    'toxicity_score'
]
 
for col in num_cols:
    if col in df_clean.columns:
        median_value = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_value)
```

```python
# Final missing value check
print("Missing values per column AFTER handling:")
print(df_clean.isnull().sum())
```

```python
df.info()
```

<h4 style="color:Black;">‚û°Ô∏èCreating derived column</h4>

<h6 style="color:Blue;">Total_Engagement</h6>

```python
# Create Total Engagement: likes + shares + comments
df_clean['Total_Engagement'] = (
    df_clean['likes_count'] +
    df_clean['shares_count'] +
    df_clean['comments_count']
)
 
# Recompute Engagement Rate if needed: Total Engagement / Impressions
# (Avoid division by zero)
df_clean['Engagement_Rate_Calc'] = np.where(
    df_clean['impressions'] > 0,
    df_clean['Total_Engagement'] / df_clean['impressions'],
    0
)
```

<h6 style="color:Blue;">Engagement_Category</h6>

```python
# Create Engagement Category based on calculated engagement rate
df_clean['Engagement_Category'] = pd.cut(
    df_clean['Engagement_Rate_Calc'],
    bins=[-0.01, 0.01, 0.05, 1.0],
    labels=['Low', 'Medium', 'High']
)
 
print("Engagement Category value counts:")
print(df_clean['Engagement_Category'].value_counts())
```

```python
# Check missing values
print("Missing values per column:")
print(df_clean.isnull().sum())
```

```python
# Fill missing text fields with 'Unknown'
df_clean['Engagement_Category'] = (
    df_clean['Engagement_Category']
    .astype('string')
    .fillna('Unknown')
)
```

```python
# Check missing values
print("Missing values per column:")
print(df_clean.isnull().sum())
```

<h4 style="color:Black;">‚û°Ô∏èCorrecting Data Types</h4>

```python
# Convert timestamp to datetime
df_clean['timestamp'] = pd.to_datetime(
    df_clean['timestamp'],
    format='mixed',
    errors='coerce'
).dt.date
```

```python
# Ensure numeric columns are proper numeric types
numeric_force_cols = [
    'sentiment_score', 'toxicity_score',
    'likes_count', 'shares_count', 'comments_count',
    'impressions', 'engagement_rate',
    'user_past_sentiment_avg', 'user_engagement_growth',
    'buzz_change_rate'
]
 
for col in numeric_force_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
# Convert some columns to category type (improves memory & performance)
category_cols = [
    'day_of_week', 'platform', 'location', 'language',
    'topic_category', 'sentiment_label', 'emotion_type',
    'brand_name', 'product_name', 'campaign_name', 'campaign_phase'
]
 
for col in category_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype('category')
```

```python
print("DATA TYPES AFTER CORRECTION:")
print(df_clean.dtypes)
```

<h4 style="color:Black;">‚û°Ô∏èStandardizing Text Columns</h4>

```python
# Standardize some key text columns: lowercase and strip spaces
 
standard_text_cols = [
    'platform', 'day_of_week', 'topic_category',
    'sentiment_label', 'emotion_type',
    'brand_name', 'product_name', 'campaign_name', 'campaign_phase'
]
 
for col in standard_text_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype(str).str.strip().str.lower()
```

```python
# check unique values after standardization
if 'platform' in df_clean.columns:
    print("Unique platforms after standardization:")
    print(df_clean['platform'].unique())
```

<h4 style="color:Black;">‚û°Ô∏èFiltering Data</h4>

<h6 style="color:Blue;">High Engagement Posts</h6>

```python
#Filter posts with High engagement
high_engagement_posts = df_clean[df_clean['Engagement_Category'] == 'High']
print("High engagement posts count:", high_engagement_posts.shape[0])
high_engagement_posts.head()
```

<h6 style="color:Blue;">Instagram Post Count</h6>

```python
# Filter posts from a specific platform (e.g., Instagram)
if 'platform' in df_clean.columns:
    instagram_posts = df_clean[df_clean['platform'] == 'instagram']
    print("Instagram posts count:", instagram_posts.shape[0])
    instagram_posts.head()
```

<h6 style="color:Blue;">Negative Sentiment Posts</h6>

```python
# Filter posts with negative sentiment
if 'sentiment_label' in df_clean.columns:
    negative_posts = df_clean[df_clean['sentiment_label'] == 'negative']
    print("Negative sentiment posts count:", negative_posts.shape[0])
    negative_posts.head()
```

<h4 style="color:Black;">‚û°Ô∏èAggregating Data</h4>

<h6 style="color:Blue;"> Average Engagement Rate by Platform</h6>

```python
if 'platform' in df_clean.columns:
    print("\nAverage Engagement Rate by Platform:")
    print(df_clean.groupby('platform')['Engagement_Rate_Calc'].mean())
```

<h6 style="color:Blue;"> Average Total Engagement by Topic Category</h6>

```python
if 'topic_category' in df_clean.columns:
    print("\nAverage Total Engagement by Topic Category:")
    print(df_clean.groupby('topic_category')['Total_Engagement'].mean())
```

```python
if 'sentiment_label' in df_clean.columns:
    print("\nAverage Engagement Rate by Sentiment Label:")
    print(df_clean.groupby('sentiment_label')['Engagement_Rate_Calc'].mean())
```

<h6 style="color:Blue;">Distribution of Engagement Rate ‚Üí Are most posts highly engaging or not?</h6>

```python
plt.figure(figsize=(8,5), facecolor="lightblue")

# Create engagement ranges
df['EngagementGroup'] = pd.cut(
    df['Engagement Rate'],
    bins=[0, 0.01, 0.05, 0.1, 0.2, 1],
    labels=["0‚Äì1%", "1‚Äì5%", "5‚Äì10%", "10‚Äì20%", "20%+"]
)

sns.countplot(x='EngagementGroup', data=df, color="blue", edgecolor="darkgreen", width=0.5)

plt.title("DISTRIBUTION OF ENGAGEMENT RATE GROUPS", fontsize=16, weight="bold", color="darkblue")
plt.xlabel("Engagement Rate Groups", fontsize=10, weight="bold", color="black")
plt.ylabel("Number of Posts", fontsize=10, weight="bold", color="black")

plt.gca().set_facecolor("lavender")
plt.xticks(color="brown", fontsize=8)
plt.yticks(color="brown", fontsize=8)

plt.tight_layout()
plt.show()
```

<h6 style="color:Blue;">Insights</h6>
<div style="text-align:justify;">
  <ul>
  <li><span style="font-size:18px"></span>The number of posts in the 1‚Äì5% and 5‚Äì10% engagement rate groups is very high, showing that most posts achieve moderate engagement levels.</li>
  <li><span style="font-size:18px"></span>The 5‚Äì10% group shows the highest spike, indicating this range is where posts perform the best on average.</li>
<li><span style="font-size:18px"></span>Posts in the 10‚Äì20% engagement group drop compared to the peak, suggesting that high engagement is less common.</li>
<li><span style="font-size:18px"></span>There is a further decline in the 20%+ group, though a good number of posts still achieve very high engagement, indicating the presence of strong or viral content.</li>
<li><span style="font-size:18px"></span>The 0‚Äì1% engagement group has very few posts, showing that extremely low-performing posts are rare in this dataset.</li>
